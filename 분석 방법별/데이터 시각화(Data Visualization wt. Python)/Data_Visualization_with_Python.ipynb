# Ch1. matplotlib 심화

**시각화를 잘해야 하는 이유**

1. 분석 결과를 효과적으로 공유하기 위해서

- 의사결정자들에게 얼마나 의미 있는 정보를 효과적으로 잘 전달하느냐

2. 분석가 스스로의 편의를 위해서

- 효과적으로 데이터를 파악하고 해석에 걸리는 시간을 단축시켜주는 효과

## matplotlib의 두 가지 인터페이스

- 인터페이스 (interface) = 동작 방식

1. **object-oriented (객체 지향적)**

- matplotlib의 기본 동작 원리
- 그래프를 구성하는 각각의 요소들을 하나씩 구체적으로 짚어가면서 구현
- 명확하게 어떤 객체에 어떤 동작이 가해지기를 바라는지 명시
- 세부적인 그래프의 조정에 더 유리

2. **state-based (상태 기반의)**

- object-oriented 방식에서 불필요해 보이는 코드가 조금씩 생략된 방식

**Figure와 Axes**

- object-oriented은 그림을 그릴 공간 (figure) 을 준비한 후, 그곳에 캔버스 (axes) 를 얹고, 그 캔버스 위에 그림을 그리는 방식

- Figure : 그래프를 담고 있는 최상위 객체

- Axes : Figure 안에 들어가는 하위 개념. 그래프가 들어가는 빈 영역 (캔버스). 하나의 캔버스에는 하나의 작품이 그려지는 것처럼 Axes에는 그래프를 하나만 얹을 수 있음

## (1) Object-oriented (객체 지향적)

그림 그릴 공간을 준비한 후, 캔버스를 얹고, 그 캔버스 위에 그림을 그리는 방식

````python
# 라이브러리 불러오기
import matplotlib.pyplot as plt
````

````python
fig, ax = plt.subplots()  # fig : 그림을 그릴 공간, ax = 캔버스
ax.plot([1, 2, 3, 4], [0, 0.5, 1, 0.2])  # 만들어 둔 캔버스 (ax) 에 그래프를 그림
````

## (2) state-based (상태 기반의)

이미 그래프를 그릴 캔버스는 눈앞에 있다고 가정하고 바로 그래프를 그리는 방식

````python
plt.plot([1, 2, 3, 4], [0, 0.5, 1, 0.2])
````

### 한 화면에 여러 개의 그래프 그리기

````python
# 데이터 불러오기
sns_df = pd.read_csv('/content/social_media_stocks.csv')
sns_df.head()
````

````python
sns_df['Symbol'].unique()
````

````python
sns_df.info()
````

````python
# Year 분리
sns_df['Year'] = sns_df['Date'].str.split('-', expand=True)[0]
````

````python
sns_df.head()
````

````python
# 연도별 기업별 평균 주식 종가 계산
sns_close_df = sns_df.groupby(['Year', 'Symbol'])[['Close']].mean().reset_index()
````

````python
# Meta (Facebook) 과 X (Twitter) 데이터 각각 저장
fb_stock_close = sns_close_df.query('Symbol == "FB"')
twtr_stock_close = sns_close_df.query('Symbol == "TWTR"')
````

````python
# 시각화
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))  # 1행 2열

# ax1 : Meta (Facebook) 의 연도별 평균 주식 종가
ax1.plot(fb_stock_close['Year'], fb_stock_close['Close'], color='teal', marker='o')

# ax2 : X (Twitter) 의 연도별 평균 주식 종가
ax2.plot(twtr_stock_close['Year'], twtr_stock_close['Close'], color='palevioletred', marker='o')

# ax1 그래프 커스터마이징
ax1.set_title('Yearly Average Close Price of FB')
ax1.set_xlabel('Year')
ax1.set_ylabel('Average Stock Close Price')
ax1.grid(axis='y', linestyle='--', color='lightgrey')
ax1.spines[['top', 'right']].set_visible(False)

# ax2 그래프 커스터마이징
ax2.set_title('Yearly Average Close Price of TWTR')
ax2.set_xlabel('Year')
ax2.set_ylabel('Average Stock Close Price')
ax2.grid(axis='y', linestyle='--', color='lightgrey')
ax2.spines[['top', 'right']].set_visible(False)
````

````python
# 위에서 작성한 공통된 코드를 for문을 사용하여 시각화
fig, axes = plt.subplots(1, 2, figsize=(12, 4))  # 1행 2열

# ax1 : Meta (Facebook) 의 연도별 평균 주식 종가
axes[0].plot(fb_stock_close['Year'], fb_stock_close['Close'], color='teal', marker='o')

# ax2 : X (Twitter) 의 연도별 평균 주식 종가
axes[1].plot(twtr_stock_close['Year'], twtr_stock_close['Close'], color='palevioletred', marker='o')

axes[0].set_title('Yearly Average Close Price of FB')
axes[1].set_title('Yearly Average Close Price of TWTR')

# 공통된 부분
for ax in axes:
    ax.set_xlabel('Year')
    ax.set_ylabel('Average Stock Close Price')
    ax.grid(axis='y', linestyle='--', color='lightgrey')
    ax.spines[['top', 'right']].set_visible(False)
````

### Seaborn : Figure-level과 Axes-level 그래프

- seaborn 라이브러리는 figure-level과 axes-level이라 불리는 두 가지 타입의 함수를 지원. figure-level 함수 하나에 여러 개의 axes-level 함수가 대응되는 1:N 관계

1) relplot (relational) - scatterplot, lineplot

2) displot (distribution) - histplot, kdeplot

3) catplot (categorical) - stripplot, swarmplot, boxplot, violineplot, pointplot, barplot

예)

- 산점도 시각화 : seaborn.scatterplot(), seaborn.relplot(kind="")
- 라인 그래프 시각화 : seaborn.lineplot(), seaborn.relplot(kind="")


1. figure-level 함수란, figure 자체를 생성하는 그래프 시각화 함수.

2. axes-level 함수란, axes를 생성하는 시각화 함수.
